```python
##### this finds all values for a list input given a value to check

#### this first part creates a double list of values and answers given a xml input string

import pprint

#CheckEdValueSring = '<w:r w:rsidRPr="00E23948"><w:rPr><w:rFonts w:ascii="Times New Roman" w:cs="Times New Roman" w:hAnsi="Times New Roman"/><w:i/><w:iCs/><w:sz w:val="32"/><w:szCs w:val="32"/></w:rPr><w:t xml:space="preserve">Italic, </w:t></w:r>'






CheckEdValueSringList = CheckEdValueSring.split('"')


#pprint.pprint(CheckEdValueSringList)






NewCheckEdValueSringList = []

for item in range(len(CheckEdValueSringList)):

    newItem = [item, CheckEdValueSringList[item]]
    
    NewCheckEdValueSringList.append(newItem)
    
    
#pprint.pprint(NewCheckEdValueSringList)
    
    
    
    
    
    
    
ValueNamesListFull = []
ValueNamesList = []
for evenItem in range(0,len(NewCheckEdValueSringList),2):
    ValueDescription = NewCheckEdValueSringList[evenItem][1].split(' ')[-1]
    ValueDescriptionFull = NewCheckEdValueSringList[evenItem][1]
    
    #print(ValueDescription)
    
    if '=' in ValueDescription:
        ValueNamesList.append(ValueDescription)
        ValueNamesListFull.append(ValueDescriptionFull)
        
#print(ValueNamesList)
        
    
    
    
    
    

#### this is for values
id = -1

TripleInfoList = []
for oddItem in range(1,len(NewCheckEdValueSringList),2):
    id = id + 1
    answerValues = NewCheckEdValueSringList[oddItem][1]
    
    TripleInfoList.append([ValueNamesList[id], answerValues, '+++++> ' +ValueNamesListFull[id]+' <+++++' ])
    
    
pprint.pprint(TripleInfoList)
    
    
```

```python
###### this gets a list of possible answers for the values given (if the xml input is big) or just the input asked

valueToCheck = 'w:val='

AnswerList = []
for item in TripleInfoList:
    if item[0] == valueToCheck:
        AnswerList.append([item[1],item[2]])
        
pprint.pprint(AnswerList)

```
