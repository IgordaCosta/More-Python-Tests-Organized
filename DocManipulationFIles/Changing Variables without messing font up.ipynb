```python
##### This is unsing a for loop for changing table variables

from docx import Document
import pprint

Location = r'C:\Users\IgorDC\Desktop\PydocTest' +'\\'


DocName = Location + 'DifferentLineText5.docx' 

document = Document(docx = DocName)


# help(document)

# p = document.paragraphs[0]

# help(document)

t = document.tables


rowN = len(t[0].rows)

columnN = len(t[0].columns)



for row in range(rowN):
    for col in range(columnN):

        p = t[0].cell(row,col).paragraphs[0]

        rr = p

        saveTextr = rr.text

        rrr = rr.runs[0]

        saveTextr = rrr.text

        word = saveTextr +' new text'

        rrr.text = ''

        run = rrr.add_text(word)






document.save(Location + 'DifferentLineTextFinished5.docx')
```
