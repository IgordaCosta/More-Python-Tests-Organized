```python
###### This gets the needed data from the table for checking purposes



from docx import Document

from bs4 import BeautifulSoup

import pprint


# OriginalDocument = 'TestDocNoChangeAllBoldArialBlack2.docx'


OriginalDocument = 'DifferentLineText6DoubleTableOnly.docx'

Location = r'C:\Users\IgorDC\Desktop\PydocTest' + '\\'




LocOriginalDocument = Location + OriginalDocument


wordDocument = LocOriginalDocument


document = Document(wordDocument)


item = document.element.xml


soup = BeautifulSoup(item)


# pprint.pprint(soup)

Table = soup.find_all('w:tbl')   ##### needs find_all when there is two or more tables

print(Table)
```

```python
pprint.pprint(Table[1]) ##### when there is two or more tables
```

```python
tblpr = Table[1].find('w:tblpr') #### find_all is not needed in case of more tables 
                                 #### this are general settings for this especific table
                                 #### since Table[1] is used a find all is not needed

pprint.pprint(tblpr)
```

```python
tblgrid = Table[1].find('w:tblgrid') #### find the grid information of the table
                                     #### this gives grid settings for the tale
                                     #### since Table[1] is used a find all is not needed

pprint.pprint(tblgrid)
```

```python
tr = Table[1].find_all('w:tr') #### find all the rows in the specific table



pprint.pprint(tr)
```

```python
tc = tr[3].find_all('w:tc')    ###### this is for finding each column information for a specific row

pprint.pprint(tc)
```

```python
pprint.pprint(tc[0])
```

```python
tcpr = tc[0].find_all('w:tcpr')   ###### this is a per column data information for a specific row column combination

pprint.pprint(tcpr)
```

```python
p = tc[0].find_all('w:p')   #### this is for finding each paragraph in the column row combination

pprint.pprint(p)
```

```python
p[0].find_all('w:ppr')    ##### this needs to stay per paragraph
```

```python
pt = p[0].find_all('w:r')   ##### this needs to stay per paragraph

pprint.pprint(pt)
```
