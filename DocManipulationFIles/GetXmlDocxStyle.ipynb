```python
AllTag = soup.find_all('p')

pprint.pprint(AllTag)
```

```python
def f(string, match):
    string_list = string.split()
    match_list = []
    for word in string_list:
        if match in word:
            match_list.append(word)
    return match_list



ParaIds = []
for Tag in AllTag:
    TagId0 = f(str(Tag), 'paraId')[0]
    TagId1 = TagId0.split('"')
    if TagId1[-1] =='':
        TagId=TagId1[-2]
    else:
        TagId=TagId1[-1]
        
        
    ParaIds.append(TagId)
    
pprint.pprint(ParaIds)
```
