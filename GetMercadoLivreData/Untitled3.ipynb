```python

try:
    # Python 2
    from urllib2 import urlopen
except ImportError:
    from urllib.request import urlopen

from bs4 import BeautifulSoup

import pprint



def GetMercadoLivreMainSearchData(url, PageNumber, CurrentDate):
    try:
        response = urlopen(url)
    
    except Exception as e:
        print(e)
        if str(e) == 'HTTP Error 404: Not Found':
            return 'PageNotFound: Done'
    

    soup = BeautifulSoup(response)
    
 
    classUsed = 'ui-search-layout__item'
    
    
    # Gets All items
    MainListOfRows = soup.find_all("li", {"class": classUsed})

    
    AllItemDataList = []
    for Row in range(len(MainListOfRows)):
        RowToCheck = MainListOfRows[Row]

        # for WebItemLocation
        SiteLocationclassUsed ='ui-search-item__group--title'
        SiteLocationdiv = RowToCheck.find_all("div", {"class": SiteLocationclassUsed})

        # for ItemImage
        ImageclassUsed = 'slick-slide slick-active'
        Imagediv = RowToCheck.find_all("div", {"class": ImageclassUsed})

        #This is for text
        TextclassUsed = 'ui-search-item__title'
        TextH2= RowToCheck.find_all("h2", {"class": TextclassUsed})

        #This is for Item Price
        PriceclassUsed = 'ui-search-price--size-medium'
        Pricediv = RowToCheck.find_all("div", {"class": PriceclassUsed})


        #####=======
        #Results of the Above By Order       
        SiteLocation = SiteLocationdiv[0].find('a')['href'] 

        ItemImage0 = Imagediv[0]
        ItemImage = ItemImage0.find('img')['data-src']

        TextH20 = TextH2[0]
        TextH2 = TextH20.text

        
        PriceCssError = False
        try:
            Price0 = Pricediv[0]
        except IndexError:
            PriceCssError = True
            
        if PriceCssError:
            PriceclassUsed = 'ui-search-item__pds-best-price'
            Pricediv = RowToCheck.find_all("div", {"class": PriceclassUsed})
            Price0 = Pricediv[0]

        Price1 = Price0.find("span", {"class": 'price-tag-fraction'})
        Price2 = Price1.text
        Price = ''.join(Price2.split('.'))

        
#         ItemDataList = [PageNumber, Row, SiteLocation, ItemImage, TextH2, Price]

        ItemDataList = [SiteLocation, ItemImage, TextH2, Price, CurrentDate]
        AllItemDataList.append(ItemDataList)

        
    return AllItemDataList

```

```python




def AppendNewDb(Database,TableName,ItemsToAddList, columns, SortValues):

    df = pandas.DataFrame(ItemsToAddList, columns = columns) 

    other = GetSqliteIntoDB(Database,TableName,SortValues)
#     print(other)
    if str(other) == 'tableDoesNotExist':
        return df
    
    else:
        df = df.append(other, ignore_index=True, verify_integrity=False, sort=False)

        df.sort_values(by=SortValues,  ascending=False,  inplace=True)
        try:
            df.drop('index',axis='columns', inplace=True)
        except:
            pass

        newIndex = list(range(len(df)))

        df['ItemOrder'] = newIndex

        df = df.set_index('ItemOrder')

        return df
```

```python

def DropSqliteTable(Database,TableName):
    conn = sqlite3.connect(Database)

    #Creating a cursor object using the cursor() method
    cursor = conn.cursor()

    #Doping EMPLOYEE table if already exists
    try:
        cursor.execute("DROP TABLE " + TableName )
        returnValue = 'TableDropedSucessfully'
    except:
        returnValue = 'TableDoesNotExist'



    #Commit your changes in the database
    conn.commit()

    #Closing the connection
    conn.close()
    
    return returnValue
```
