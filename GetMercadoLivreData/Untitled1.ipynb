```python
try:
    # Python 2
    from urllib2 import urlopen
except ImportError:
    from urllib.request import urlopen

from bs4 import BeautifulSoup

import pprint



url = 'https://eletronicos.mercadolivre.com.br/drones-e-acessorios/novo/drone_DisplayType_LF_OrderId_PRICE*DESC_PriceRange_600-1100_NoIndex_True'

    
response = urlopen(url)

soup = BeautifulSoup(response)

classUsed = 'ui-search-price--size-medium'

mydivs = soup.find_all("div", {"class": classUsed})


# print(mydivs)

# print(len(mydivs))
#This is for pricetag

ItemImageList = []
for Item in range(len(mydivs)):
    ItemImage0 = mydivs[Item]
    
#     print(ItemImage0)
#     ItemImage = ItemImage0.text
    ItemImage1 = ItemImage0.find("span", {"class": 'price-tag-fraction'})
    
    ItemImage2 = ItemImage1.text
    
    ItemImage = ''.join(ItemImage2.split('.'))
    
#     print(ItemImage)
    ItemImageList.append(ItemImage)


pprint.pprint(ItemImageList)

print(len(ItemImageList))
```
